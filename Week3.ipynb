```
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(-3, 3)
x2 = np.linspace(-4, 4)
#x3 = np.linspace(-4, 4)



y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

a = np.random.normal(0, 1, 1000)
a = np.round(a, 2)
mean = np.mean(a)
std = np.std(a)
b = Counter(a)

print('평균 : ' + str(mean))
print('표준편차 : ' + str(std))

plt.title("Normal Distribution (loc=0, scale=1, sample=1000)")
for key, value in b.items():
  plt.scatter(key,value, color="b")

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

a = np.random.normal(100, 5, 100000)
a = np.round(a, 2)
mean = np.mean(a)
std = np.std(a)

a = a - mean
a = a/std
a = np.round(a, 2)

mean = np.mean(a)
std = np.std(a)

b = Counter(a)

print('평균2 : ' + str(mean))
print('표준편차2 : ' + str(std))

plt.title("Standard Normal Distribution (loc=100, scale=5, sample=1000)")
for key, value in b.items():
  plt.scatter(key,value, color="b")

plt.show()
```

```
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt

def _function(x):
  return x**3-1



sol = optimize.root_scalar(f=_function, bracket=[-10, 10], method="brentq")
print('정답: ' + str(sol.root))

#plt.title("Standard Normal Distribution (loc=100, scale=5, sample=1000)")
for i in range(-10, 10): 
  plt.scatter(i,_function(i), color="r")   

plt.show()
```
